## Arquitectura

Pytorch sigue una línea más orientada a objetos, lo cual nos da más libertad (o al menos más facilidad) para realizar operaciones en los pasos intermedios. Solo debemos extender de "nn.Module", llamar a la inicialización de "super" e implementar el método "forward" que es el que se ejecutará cuando utilicemos la red (ya sea en el entrenamiento o la predicción).

```python
class Autoencoder(nn.Module):
    def __init__(self, img_shape=(28, 28)):
        super(Autoencoder, self).__init__()
        """ Se definen las capas que tiene nuestra red, y como esta configurada
            cada una de ellas. Aquí no importa el orden en que las definamos, ya que
            el orden está dado en como las aplicamos a la entrada en 'forward' """
        self.out_s = np.array(img_shape)/8

        # Capas para etapa de Codificación
        self.c1 = nn.Conv2d( 1, 10, 3, padding=1, stride=1 )
        self.c2 = nn.Conv2d( 10, 20, 3, padding=1, stride=1 )
        self.c3 = nn.Conv2d( 20, 40, 3, padding=1, stride=1 )

        # Capas para etapa de Decodificación
        self.ct1 = nn.ConvTranspose2d( 40, 20, 3, padding=1, stride=2 )
        self.ct2 = nn.ConvTranspose2d( 20, 10, 3, padding=1, stride=2 )
        self.ct3 = nn.ConvTranspose2d( 10, 1, 3, padding=1, stride=2 )

    def encode( self, x ):
        """ Codificación. Aquí especificamos en que orden se aplican las capas"""
        x = nn.ReLU(True)( self.c1( x ) )
        x = nn.MaxPool2d( kernel_size=2, stride=2 )( x )
        x = nn.ReLU(True)( self.c2( x ) )
        x = nn.MaxPool2d( kernel_size=2, stride=2 )( x )
        x = nn.ReLU(True)( self.c3( x ) )
        x = nn.MaxPool2d( kernel_size=2, stride=2 )( x )
        return x

    def decode( self, code ):
        """ Decodificación. Aquí especificamos en que orden se aplican las capas"""
        out_s = self.out_s * 2
        x = self.ct1( code, output_size=out_s )
        x = nn.ReLU( True )( x )
        out_s = out_s*2
        x = self.ct2( x, output_size=out_s )
        x = nn.ReLU( True )( x )
        out_s = out_s*2
        x = self.ct3( x, output_size=out_s )
        x = nn.Tanh()( x )
        return x
    
    def forward(self, x):
        """ Separamos el forward en Encode y Decode. No es necesario,pero lo
            hacemos para que quede más claro el funcionamiento del Autoencoder """
        code = self.encode( x )
        rec = self.decode( code )
        return rec
```

```python
model = Autoencoder( img_shape=(64, 64)).double()
print(model)
```

```python
# Pytorch trabaja con Tensores, no con numpy arrays
# así que los convertimos con una función utilitaria propia (ver 'utiles')
x_train_raw, x_train_dl = to_tensor(x_train, b_size=50)
x_test_raw, x_test_dl = to_tensor(x_test, b_size=50)
```

```python
criterion = nn.MSELoss()                                      # Métrica utilizada para obtener el gradiente
optimizer = torch.optim.Adam( model.parameters(), lr=0.0001 ) # Método de ptimización de los pesos
n_epochs = 1                                                  # Cantidad de epochs (tiene 1 porque ya  lo 
for e in range( n_epochs ):                                   # entrené y tengo los pesos guardados) 
    for d in x_train_dl:
        img = d[0]
        img = Variable(img)

        optimizer.zero_grad()               # Inicializamos los gradientes (normalmente se hace por cada batch)
        output = model(img)
        loss = criterion(output, img)       # Calculamos la función de pérdida 
        loss.backward()                     # Calculamos el gradiente para cada peso
        optimizer.step()                    # Actualizamos los pesos de la red

    print("epoch [{}/{}], loss:{:.4f}".format(e + 1, n_epochs, loss.item() ) )
```

torch.save(model.state_dict(), "pesos/pytorch_ae_conv.h5")
