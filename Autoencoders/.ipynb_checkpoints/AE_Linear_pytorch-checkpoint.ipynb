```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from utiles import *
```

```python
import torch
import torchvision
from torch import nn
from torch.autograd import Variable
```

### Cargamos y vemos como son los datos

```python
data = data_loader("data/lfwcrop_grey", shape=(56, 56))
x_train, x_test = train_test_split( data, test_size=.25, random_state=15)
data.shape, x_train.shape, x_test.shape
```

```python
plot_n_faces(data, data, 2)
```

```python
class Autoencoder(nn.Module):
    def __init__(self, img_shape=(56, 56)):
        super(Autoencoder, self).__init__()
        """ Se definen las capas que tiene nuestra red, y como esta configurada
            cada una de ellas. Aquí no importa el orden en que las definamos, ya que
            el orden está dado en como las aplicamos a la entrada en 'forward' """
        w, h = img_shape
        self.l1 = nn.Linear( w*h, 512 )
        self.l2 = nn.Linear( 512, 256 )
        self.l3 = nn.Linear( 256, 64 )

        self.ld1 = nn.Linear( 64, 256 )
        self.ld2 = nn.Linear( 256, 512 )
        self.ld3 = nn.Linear( 512, w*h )

    def encode( self, data ):
        """ Codificación. Aquí especificamos en que orden se aplican las capas"""
        x = self.l1( data )
        x = nn.ReLU( True )( x )
        x = self.l2( x )
        x = nn.ReLU( True )( x )
        x = self.l3( x )
        return x

    def decode( self, code ):
        """ Decodificación. Aquí especificamos en que orden se aplican las capas"""
        x = self.ld1( code )
        x = nn.ReLU( True )( x )
        x = self.ld2( x )
        x = nn.ReLU( True )( x )
        x = self.ld3( x )
        x = nn.Tanh()( x )
        return x
    
    def forward(self, x):
        """ Separamos el forward en Encode y Decode. No es necesario,pero lo
            hacemos para que quede más claro el funcionamiento del Autoencoder """
        code = self.encode( x )
        rec = self.decode( code )
        return rec
```

```python
model = Autoencoder( img_shape=(56, 56))
model.double()
```

```python
# Pytorch trabaja con Tensores, no con numpy arrays
# así que los convertimos con una función utilitaria propia (ver 'utiles')
x_train_raw, x_train_dl = to_tensor(x_train, b_size=50)
x_test_raw, x_test_dl = to_tensor(x_test, b_size=50)
```

```python
optimizer = torch.optim.Adam( model.parameters(), lr=0.0001 )
criterion = nn.MSELoss()

n_epochs = 1
for e in range( n_epochs ):
    for d in x_train_dl:
        optimizer.zero_grad()
        img = d[0]
        img = img.view(d[0].size(0), -1)
        img = Variable(img)

        output = model(img)
        loss = criterion(output, img)
        loss.backward()
        optimizer.step()

    print("epoch [{}/{}], loss:{:.4f}".format(e + 1, n_epochs, loss.item()))
```

```python
torch.save(model.state_dict(), "pesos/pytorch_ae_linear.h5")
```
