## Keras

```python
from keras.models import Sequential
from keras.layers import Conv2D, Conv2DTranspose, MaxPooling2D, UpSampling2D, Reshape, Dense, Flatten, SeparableConv2D
```

## Arquitectura del Modelo
* Reshape: Keras espera inputs de tamaño ancho*alto*num_canales. En el caso de imágenes (r,g,b) serían 3, pero como estamos viendo imágenes en escala de grises, solo tienen 1 canal.
* Conv2D: Realiza las convoluciones con la cantidad y tamaño de kernel especificado. Ver documentación para más información.
* MaxPooling2D: Da a la siguiente capa solo los filtros que más se activaron. El valor "2" el tamaño de la ventana en que se aplica ( en este caso, ventanas de 2*2), por lo cual la capa siguiente tiene un shape = 1/2 shape input.

```python
model = Sequential()

model.add( Reshape( (64, 64, 1), input_shape=(64, 64) ))
model.add( Conv2D( filters=16, kernel_size=5, activation="relu", padding="same") )
model.add( MaxPooling2D( 2 ))
model.add( Conv2D( filters=32, kernel_size=3, activation="relu", padding="same") )
model.add( MaxPooling2D( 2 ))
model.add( Conv2D( filters=64, kernel_size=3, activation="relu", padding="same") )
model.add( MaxPooling2D( 2 ))

model.add( Conv2D( filters=128, kernel_size=3, activation="relu", padding="same") )
model.add( UpSampling2D( 2 ))

model.add( Conv2D( filters=64, kernel_size=3, activation="relu", padding="same") )
model.add( UpSampling2D( 2 ))
model.add( Conv2D( filters=32, kernel_size=3, activation="relu", padding="same") )
model.add( UpSampling2D( 2 ))
model.add( Conv2D( filters=16, kernel_size=3, activation="relu", padding="same") )
model.add( Conv2D( filters=1, kernel_size=3, activation="tanh", padding="same") )


model.add( Reshape( (64, 64), input_shape=(64, 64, 1) ) )
```

La función "summary" nos ofrece un vistazo general de como quedó estructurada nuestra red (incluyendo información del shape de las salidas de cada capa en ancho\*alto\*cant_filtros). Notar como en las capas más internas, se tienen 128 filtros de shape 8\*8 en vez de la imagen de 64\*64

```python
model.summary()
```

```python
model.compile( loss="mse", optimizer="adam", metrics=["mse"])
```

```python
# History es un diccionario en el que queda guardado el progreso del modelo en cada epoch (loss, val_loss, etc)
history = model.fit(x_train, x_train, verbose=1,
              batch_size=50,
              epochs=1,
              validation_split=.2)
```

## Y si las imágenes tienen ruido?

```python
# Seteamos el % de ruido que queremos
n_ratio = .15
# Agregamos ruido con una distribución normal a las imágenes
xn_test = x_test + n_ratio*np.random.normal(0, 1, x_test.shape)
# Con "clip" nos aseguramos de que los valores no esten por arriba de 1 (blanco) ni por debajo de 0 (negro)
xn_test = np.clip(xn_test, a_max=1, a_min=0)
```

```python
new_p = model.predict( xn_test[:100] )
```

```python
plot_n_faces(xn_test[:100], x_test[:100], n=5, predicted=new_p)
```
